## 4. Data Cleaning

```python
# Initial exploration showed 100% null for this column; Data found in another file. 
df.drop(columns=["Product Description"], inplace=True)
```

```python
# Initial exploration showed ~85% null; Location granularity also unnecessary.
df.drop(columns=["Order Zipcode"], inplace=True)
```

```python
# Standardized column names
df.columns = (df.columns.str.strip().str.lower().str.replace(" ", "_").str.replace("(", "").str.replace(")", ""))
df.columns
```

```python
# Dates did not have an actual datetime data type.
df['order_date_dateorders'] = pd.to_datetime(df['order_date_dateorders'])
df['shipping_date_dateorders'] = pd.to_datetime(df['shipping_date_dateorders'])
```

## 5. Feature Engineering 

```python
# Recreated Shipping Delay column to ensure accuracy and control of the metric.
df['shipping_delay'] = (df['shipping_date_dateorders'] - df['order_date_dateorders']).dt.days
```

```python
df[['shipping_delay', 'days_for_shipping_real']].head(10)
```

```python
df.drop(columns=['days_for_shipping_real'], inplace=True)
```

## 6. Operational Analysis 

### Shipping Performance 

```python
df.groupby('shipping_mode')['shipping_delay'].mean().sort_values()
```

Insights:
- Same-day shipping is the fastest option, as expected.
- First class provides a significant improvement over slower methods.
- Second Class and Standard Class have nearly identical delivery times(~4 days), suggesting minimal performance difference despite being different service tiers.

This indicates possible inefficiencies or a lack of differentiation between mid-tier shipping options. 

### Distribution of Delays

```python
df['shipping_delay'].describe()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['shipping_delay'], bins=30, kde=True)
plt.title("Distribution of Shipping Delays")
plt.xlabel("Days")
plt.ylabel("Frequency")
plt.show()
```

The distribution shows a clear concentration around 2-day shipping delays, which correlate with First Class shipping and suggest that a significant portion of orders are fulfilled using mid-tier shipping rather than the other shipping methods.

```python
# Verifying that there are no negative and/or extreme values not showing properly on the visual. 
df['shipping_delay'].min(), df['shipping_delay'].max()
```

```python
# Delivery performance across regions
df.groupby('order_region')['shipping_delay'].mean().sort_values(ascending=False)
```

While some regions, such as Central, East, and West Africa, exhibit higher average delays, all regions fall within a 3-3.5-day delay range, indicating relatively consistent performance. 

This suggests that shipping inefficiencies are not region-specific but rather systemic. 

```python
# Checking region consistency on top of shipping speed. 
df.groupby('order_region')['shipping_delay'].std().sort_values(ascending=False)
```

### Late Delivery Risk

```python
df.groupby('late_delivery_risk')['shipping_delay'].mean()
```

The average delay between late and on-time deliveries is not extreme, but late deliveries are about 1.3 days slower. At scale, this results in a meaningful operational delay and likely contributes to customer dissatisfaction and higher fulfillment costs. 

```python
df['late_delivery_risk'].value_counts()
```

```python
df['late_delivery_risk'].value_counts(normalize=True)
```

Approximately 55% of all shipments are classified as late deliveries, indicating that late fulfillment is the norm rather than the exception. This further suggests that there are systemic inefficiencies in fulfillment and logistics rather than isolated incidents.

## Profit & Business Impact

### Profit Impact of Delays

```python
df.groupby('late_delivery_risk')['order_profit_per_order'].mean()
```

```python
df.groupby('late_delivery_risk')['sales'].mean()
```

Late deliveries are associated with approximately a \\$0.78 drop in profit per order compared to on-time orders and also show a larger drop of \\$0.95 in sales. 

Given that about 55% of all orders are classified as late (~99,000 orders), this corresponds to an estimated \\$77K in reduced profit across the dataset.  

While this does not prove that late deliveries directly cause the profit decrease, the association suggests that delivery performance may have some meaningful impact at scale. 

### Profits Across Other Features

```python
df.groupby('shipping_mode')['order_profit_per_order'].mean()
```

Insights: 
- First-class shipping displays the highest profit amongst the current shipping methods. 
- Same-day shipping displays the lowest profit, and when considering the likely increased operational cost, may not be profitable at scale under current operating conditions.
- Second-class and Standard class display nearly identical metrics, with Standard shipping leading by ~\\$0.70. Not only does Second-class have a similar delay speed compared to Standard, but it also generates less profit.  

```python
df.groupby('order_region')['order_profit_per_order'].mean().sort_values(ascending=False)
```

Shipping speed and consistency do not directly predict profitability. While some high-performing regions demonstrate both strong consistency and high profit (e.g., Canada), others deviate from this pattern, indicating that profitability is influenced by multiple operational and regional factors beyond shipping performance alone.

### Late Deliveries Per Item Category

```python
df.groupby('category_name')['late_delivery_risk'].mean().sort_values(ascending=False)
```

Item category does not have a direct effect on shipping delays, since many categories revolve around sports equipment. However, the categories with the highest shipping delays tend to include products that are larger, heavier, and awkwardly shaped, which are likely more warehouse-intensive and increase operational complexity.
